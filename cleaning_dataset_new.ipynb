```python
import numpy as np

# df4 already has these columns:
# ['batsman', 'total_runs', 'out', 'numberofballs', 'average', 'strikerate']

# Make sure total_runs & out are numeric (values won't change, only dtype)
df4['total_runs'] = pd.to_numeric(df4['total_runs'], errors='coerce')
df4['out']        = pd.to_numeric(df4['out'],        errors='coerce')

#  Overwrite ALL values in column 'average' using the formula:
#     average = total_runs / out
#     (if out == 0 -> set as NaN or any value you prefer)

df4['average'] = df4.apply(
    lambda row: row['total_runs'] / row['out'] if row['out'] != 0 else np.nan,
    axis=1
)

# Round to 2 decimals
df4['average'] = df4['average'].round(2)
```

```python
df4['average'] = df4['average'].fillna(0)
```

```python
df4['strikerate'] = df4.apply(
    lambda x: (x['total_runs'] / x['numberofballs']) * 100 
              if x['numberofballs'] != 0 else None,
    axis=1
)
```

```python
df4['average'] = df4['average'].round(2)
df4['strikerate'] = df4['strikerate'].round(2)
```

```python


matches = pd.read_csv(r"D:\new_cricket_prediction\updated_match.csv")   # or matches.csv
teams   = pd.read_csv(r"D:\new_cricket_prediction\updated_teams.csv")

```

```python
matches
```

```python
matches['venue'] = (
    matches['venue']
    .astype(str)
    .str.strip()                   # Remove extra spaces at start/end
    .str.replace(r'\s+', ' ', regex=True)  # Replace multiple spaces with 1
    .str.upper()                   # Make ALL UPPER CASE
)
```

```python
for col in ['team1', 'team2', 'winner', 'toss_winner']:
    matches[col] = (
        matches[col]
        .astype(str)
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
        .str.upper()
    )

# Create franchise columns (keep original team1/team2 as history)
matches['team1_franchise']       = matches['team1'].map(team_map).fillna(matches['team1'])
matches['team2_franchise']       = matches['team2'].map(team_map).fillna(matches['team2'])
matches['winner_franchise']      = matches['winner'].map(team_map).fillna(matches['winner'])
matches['toss_winner_franchise'] = matches['toss_winner'].map(team_map).fillna(matches['toss_winner'])
```

```python
unmapped_teams = set(matches['team1'].unique()) - set(teams['Team_Name'].unique())
print("Unmapped team names:", unmapped_teams)
```

```python
matches.duplicated().sum()
```

```python
matches['venue'] = matches['venue'].astype(str).str.strip()
print("Missing venue count:", (matches['venue'] == '' ).sum() + matches['venue'].isna().sum())
# Treat empty strings as NaN
matches['venue'] = matches['venue'].replace(['', 'nan', 'NaN'], np.nan)

# If venue is NaN but city exists, fill with city
if 'city' in matches.columns:
    matches.loc[matches['venue'].isna() & matches['city'].notna(), 'venue'] = (
        matches.loc[matches['venue'].isna() & matches['city'].notna(), 'city']
    )

# Remaining NaN venues  UNKNOWN VENUE
matches['venue'] = matches['venue'].fillna('UNKNOWN VENUE')
```

```python
print(matches['result'].value_counts(dropna=False))
print("Missing winners:", matches['winner'].isna().sum())
```

```python
# Make sure result is clean upper text
matches['result'] = matches['result'].astype(str).str.strip().str.upper()
matches['winner_franchise'] = matches['winner_franchise'].astype(str).str.strip()

def clean_winner(row):
    if row['result'] in ['NO RESULT', 'TIE']:
        return 'NO RESULT'
    # if winner missing but result is "NORMAL", leave as NaN to investigate
    if (row['winner_franchise'] == 'NAN' or row['winner_franchise'] == '') and row['result'] == 'NORMAL':
        return np.nan
    return row['winner_franchise']

matches['winner_clean'] = matches.apply(clean_winner, axis=1)
```

```python
print("Duplicates:", matches.duplicated().sum())
print("Missing venue:", matches['venue'].isna().sum())
print("Missing winner_clean:", matches['winner_clean'].isna().sum())
print("Unique franchises (team1):", matches['team1_franchise'].unique())
print("Unique venues:", len(matches['venue'].unique()))
```
